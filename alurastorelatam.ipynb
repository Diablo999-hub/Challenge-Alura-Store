### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
suma_precio_tienda1 = sum(tienda['Precio'])
suma_precio_tienda2 = sum(tienda2['Precio'])
suma_precio_tienda3 = sum(tienda3['Precio'])
suma_precio_tienda4 = sum(tienda4['Precio'])
print(f'Las ventas totales de la tienda 1: {suma_precio_tienda1}')
print(f'Las ventas totales de la tienda 2: {suma_precio_tienda2}')
print(f'Las ventas totales de la tienda 3: {suma_precio_tienda3}')
print(f'Las ventas totales de la tienda 4: {suma_precio_tienda4}')
```

```python
tiendas = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']
ventas = [suma_precio_tienda1, suma_precio_tienda2, suma_precio_tienda3, suma_precio_tienda4]
plt.bar(tiendas,ventas,color='skyblue')
plt.title('Ventas por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
lista_mas_vendidos1 = []
for i in range(len(tienda['Calificación'])):
  mas_vendido = tienda['Producto'][i],tienda['Precio'][i], tienda['Categoría del Producto'][i]
  lista_mas_vendidos1.append(mas_vendido)


conteo_categorias = {}

# Recorrer la lista y contar
for producto in lista_mas_vendidos1:
    categoria = producto[2]
    if categoria in conteo_categorias:
        conteo_categorias[categoria] += 1
    else:
        conteo_categorias[categoria] = 1

# Mostrar conteo por categoría
print(conteo_categorias)

# Encontrar la categoría más vendida
categoria_mas_vendida = max(conteo_categorias, key=conteo_categorias.get)
print("Categoría más vendida:", categoria_mas_vendida)
```

```python
productos = list(conteo_categorias.keys())
num_productos = list(conteo_categorias.values())
plt.pie(num_productos,labels=productos)
plt.title('Categorias Más vendidas')
# plt.title('Ventas por tienda')
# plt.xlabel('Tienda')
# plt.ylabel('Ventas')
# plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
dict_productos = {}
for producto in tienda['Producto']:
  if producto not in dict_productos:
    dict_productos[producto] = 1
  else:
    dict_productos[producto] += 1
dict_productos.items()
```

```python
# Obtener valor máximo y mínimo de ventas
max_ventas = max(dict_productos.values())
min_ventas = min(dict_productos.values())

# Productos más vendidos
mas_vendidos = [producto for producto, cantidad in dict_productos.items() if cantidad == max_ventas]

# Productos menos vendidos
menos_vendidos = [producto for producto, cantidad in dict_productos.items() if cantidad == min_ventas]

print("Productos más vendidos:", mas_vendidos)
print("Productos menos vendidos:", menos_vendidos)
```
